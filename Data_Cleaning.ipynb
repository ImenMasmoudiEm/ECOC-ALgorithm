```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import numpy as np
import pandas as pd
import os
```

```python
df = pd.read_csv('/content/sample_data/ecommerceDataset.csv', names = ['category','text'])
df=df.dropna()
df.head(5)
```

```python
df.category.value_counts()
```

```python
df = df.drop(df[df['category'] == 'Clothing & Accessories'].index)
```

```python
df.category.value_counts()
```

```python
df['category'] = df['category'].map({'Household':0,'Books':1,'Electronics':2})
```

```python
df.category.value_counts()
```

```python
df.to_csv('EcommerceDataset.csv')
```

```python
training_size=int(len(df['text'])*0.9)
Text = df['text']
Lables = df['category']

training_sentences = Text[0:training_size]
training_Lables = Lables[0:training_size]
testing_sentences = Text[training_size:]
testing_Lables = Lables[training_size:]

tokenizer = Tokenizer(num_words=3000, oov_token="")
tokenizer.fit_on_texts(training_sentences)
```

```python
df1 = pd.DataFrame()
def count_words(input):
  return len(input.split())
max_length=count_words(max(training_sentences,key=count_words))
max_length
df1['Word Count'] = [count_words(i) for i in training_sentences]
df1.describe()
```

```python
training_sentences = tokenizer.texts_to_sequences(training_sentences)
training_sentences = pad_sequences(training_sentences, maxlen=200, padding='post', truncating='post')

testing_sentences = tokenizer.texts_to_sequences(testing_sentences)
testing_sentences = pad_sequences(testing_sentences, maxlen=200, padding='post', truncating='post')
```
